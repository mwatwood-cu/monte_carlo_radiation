## Part 3 - 3D Calculations

Following a similar pattern to the 2-stream model, the complete 3-D model is divided up into helper functions, a `track_photon` method that is used as the crux of the code, and then this is combined in a `full_sim` function that is able to run all desired photons and give percentage results.

Note here that three different random distributions are used. First in the `draw_random_tau` function, the exponential distribution is used to calculated an optical depth, $\tau$ travelled. Next in the `draw_random_angles` function the variable $\theta$ is calculated from the Henyey-Greenstein distribution, and finally the variable $\phi$ is calculated from a uniform distribution and scaled by $2\pi$. 

The `make_matrix` function is used to recreate the matrix given in the text as equation (6.79) and makes use of both the $\theta$ and $\phi$ values that are created in the draw random angles function. 

Finally the `track_photon` function became slightly more complicated when compared to the previous two stream model, but note the comments in line in the code for full details on the functionality.

Now that the functions have been made we can run the simulation in a similarly small code block. Note the specific taus used to compare to Figure 6.8 in the textbook.
