### Introduction
This is a project using a Monte Carlo simulation for the movement of photons through cloud layers. This simulation uses the basic principle of mean free path and Beer's law of exponential attenuation to create a probablistic way to approximate optical path of photons. The project is broken into three parts. First, in part one the probablistic method is tested for three different distributions. Next, in part two the exponential decay distribution is used to model a two stream simulation for clouds and optical path of photons. Here both with and without absorption are used and compared to exact solutions and further a root-mean-squared error method is used to measure convergence. Finally, in part three the exponential distribution is used for optical depth with the Henyey-Greenstein distribution used to determine angle of scattering for a full three dimensional simulation. Here the result is compared qualitatively to a previous work in total reflectance, transmittance, and internal relative irradiance. Finally, not shown in this project, but linked in a Google Drive, is a self contained software application that runs a visualized version of a 3D model with total optical depth of 10.

Note, minimal code is shown in this document for brevity. Code is available in its entirety at the following GitHub Repository with figures and discussion saved inline there as well.

## Henyey-Greenstein Function

As described in our text, Fundamentals of Atmospheric Radiation, the Henyey-Greenstein phase function is a mathematical tool that is used to roughly approximate the distribution of photon scattering distribution calculated from Mie theory. This phase function is used for its favorable mathematical properties including being normalized and following the appropriate integrated relation to the asymmetry parameter. 

This is the most complicated distribution used where the uniformly distributed variable $\xi$ is transformed to our desired distribution variable $x$ using the following: $$ x = \frac{1}{2g}\left( 1+g^2 - \left( \frac{1-g^2}{1-g+2g\xi} \right)^2 \right) $$

The below code creates a modified calc_dist function that uses the g asymmetry parameter and then using the hg_phase_dist function to calculate the appropriate relationship to create the correct output.
